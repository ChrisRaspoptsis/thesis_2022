```python
from sklearn.svm import SVC
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
```

```python
# split the dataset into independent (X) and dependent (Y) (m12) which is the target value
X=df.iloc[:,1:-1].values
Y=df.iloc[:,-1].values
```

```python
classifier = SVC(kernel='rbf')
```

```python
scores = []
cv = KFold(n_splits=10, random_state=42, shuffle=True)
for train_index, test_index in cv.split(X):
    X_train, X_test, Y_train, Y_test = X[train_index], X[test_index], Y[train_index], Y[test_index]
    classifier.fit(X_train, Y_train)
    scores.append(classifier.score(X_test, Y_test))
ac_svm=np.mean(cross_val_score(classifier, X, Y, cv=10))
print(np.mean(cross_val_score(classifier, X, Y, cv=10)))
```

<font color='red'>Accuracy: 76.12%</font>
